### Text preprocessing

```python
REPLACE_HASHTAG = re.compile('[#]+[A-Za-z-_]+[A-Za-z0-9-_]')
REPLACE_AT_ENTITIES = re.compile('[@]+[A-Za-z0-9-_]+')
REPLACE_URLS = re.compile(urlmarker.WEB_URL_REGEX)
REPLACE_EMAILS = re.compile('^([a-zA-Z0-9_\-\.]+)@([a-zA-Z0-9_\-\.]+)\.([a-zA-Z]{2,5})$')
REPLACE_BY_SPACE_RE = re.compile('[/(){}\[\]\|@,;]')
REPLACE_DOLLAR_FIGURES = re.compile('[\$]{1}[\d,]+\.?\d{0,2}')
REMOVE_BAD_SYMBOLS = re.compile('[^0-9a-z #+_]')
REMOVE_STOPWORDS = set(stopwords.words('english'))

def text_prepare(text):
    """
        text: a string
        
        return: modified initial string
    """
    text = text.lower()# lowercase text
    text = REPLACE_HASHTAG.sub(' 000hashtag000 ', text) # replace hashtags with special token
    text = REPLACE_EMAILS.sub(' 000email000 ', text) # replace emails with special token
    text = REPLACE_DOLLAR_FIGURES.sub(' 000dollars000 ', text) # replace dollar amounts with special token
    text = REPLACE_AT_ENTITIES.sub(' 000at_entity000 ', text) # replace @entity with special token
    text = REPLACE_URLS.sub(' 000url000 ', text) # replace urls with special token
    text = REPLACE_BY_SPACE_RE.sub(' ', text)# replace REPLACE_BY_SPACE_RE symbols by space in text
    text = REMOVE_BAD_SYMBOLS.sub('', text) # delete symbols which are in BAD_SYMBOLS_RE from text
    text = text.replace('#', '') # delete remaining hashtags
    text = ' '.join([word for word in text.split() if word not in REMOVE_STOPWORDS]) # delete stopwords from text
    return text
```

###### Supervised models

### Custom trained vectors

```python
from sklearn.cluster import *
from sklearn.mixture import GaussianMixture
```

```python
# turn each fb post into a sentence vector
sentVecs=[]
for _, row in fb_clean.iterrows():
    vec = model_unsupervised_custom.get_sentence_vector(row.content)
    sentVecs.append(vec)
sentVecs = np.array(sentVecs)
```

```python
# # grid search for best clustering
# from sklearn.model_selection import GridSearchCV

# tuned_params = {'covariance_type': ['full'], 
#                 'n_components': [8,10,12,14,16]}

# clust_model = GridSearchCV(GaussianMixture(),
#                           tuned_params,
#                           cv=5,
#                           n_jobs=6,
#                           return_train_score=True,
#                           verbose=3)
# Results: selected 10 clusters, but the expectation didn't vary much from group to group

# df_cv_results = pd.DataFrame(clust_model.cv_results_)
# sns.barplot(x="param_n_components", y="mean_test_score", 
#             hue="param_covariance_type", 
#             data=df_cv_results, 
#             ci='sd')
# plt.legend(bbox_to_anchor=(1.05, 1), loc=2, borderaxespad=0.);

clust_model = GaussianMixture(n_components=25, covariance_type='full')
clust_model.fit(sentVecs)
```

```python
# add the group label
predicted_labels=[str(lab) for lab in clust_model.predict(sentVecs)]
group_label = pd.DataFrame(predicted_labels, columns=['group'])
df_results = pd.concat([fb_clean.reset_index(),group_label.reset_index()], axis=1)

# add the group probabilities
predicted_probs=[np.max(z) for z in clust_model.predict_proba(sentVecs)]
group_prob = pd.DataFrame(predicted_probs, columns=['group_prob'])
df_results = pd.concat([df_results.reset_index(),group_prob.reset_index()], axis=1)
df_results = df_results.drop('index',axis=1)
```

```python
from collections import Counter

def join_words(x):
    long_string = ' '.join(x)
    splitString = long_string.split()

    ctr = Counter(splitString)

    top_n_words = [ x[0] for x in ctr.most_common(12)]
    
    short_string = ', '.join(top_n_words)
    
    return(short_string)
    

df_summary = df_results.groupby(['group'])['content'].agg([len, join_words])
df_summary.columns = ['num_tweets', 'top_words']

print_default_length = pd.options.display.max_colwidth
pd.options.display.max_colwidth = 280
print(df_summary)
pd.options.display.max_colwidth = print_default_length
```

```python
# use pca or tsne to visualize in 2 dimensions
tsne = TSNE(n_components=2, 
                     angle=0.5, 
                     verbose=3, 
                     perplexity=20,
                     n_iter=500,
                     n_jobs=6)

# Get the sentence embeddings
df_sent_embedded = pd.DataFrame(tsne.fit_transform(sentVecs), columns=['x', 'y'])
```

```python
# join onto the source data
df_plot = pd.concat([df_results.reset_index(), df_sent_embedded.reset_index() ], axis=1).drop('index', axis=1)
df_plot = df_plot.astype(dict(negpost='str', x='float', y='float'))

# Plot the t-sne groupings
print(df_plot.head())
print(sns.lmplot(x='x', y='y', fit_reg=False, legend=True, hue='group', data=df_plot, size=15));
```
