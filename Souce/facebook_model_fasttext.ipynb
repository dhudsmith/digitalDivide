## Preliminaries

```python
# Essentials
import pandas as pd
import numpy as np
import itertools
from docx import Document
```

## Data Preparation

```python
dataPath = '/home/hudson/Code/DigitalDivide/Data/digital_divide_2018-04-14_36.csv'
dd_data=pd.read_csv(dataPath, encoding='utf-8', sep=',')[['Q2.2']]
```

```python
q2_2 = dd_data['Q2.2'][3:]
q2_2 = q2_2[~q2_2.isnull()]
q2_2[:10]
```

```python
# Create and save a word document
def write_word(file, text):
    doc = Document()
    doc.add_paragraph(text)
    doc.save(file)
```

```python
# downsample for testing
nsample = q2_2.shape[0]
q2_2_sampled = q2_2.sample(nsample, replace=False)

# write the files
path =  '/home/hudson/Code/DigitalDivide/Data/ForWatson/'
docxfiles=[]
for num, text in enumerate(q2_2_sampled):
    filename = 'q2_2_answer_'+str(num)+'.docx'
    fullpath = path+filename
    docxfiles.append(fullpath)
    print("Creating word file: " + fullpath)
    
    write_word(path+filename, text)
```
